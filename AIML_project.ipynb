# Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Create a linear regression object and fit the model to the training data
lr = LinearRegression()
lr.fit(X_train, y_train)

```

```python
lr.intercept_    #If Y= B0+B1*X1+B2*X2+...
#B0
```

```python
lr.coef_
#(B1,B2...,B5)
```

# performance of Linear regression


```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np


# make predictions
y_pred = lr.predict(X_test)

# calculate R2 value
r2 = r2_score(y_test, y_pred)

# calculate RMSE value
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("R2 value:", r2)
print("RMSE value:", rmse)
```

```python
# plot the actual vs predicted values
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Yield')
plt.ylabel('Predicted Yield')
plt.title('Actual vs Predicted Yield')
plt.show()
```

# Kernel Regression

```python
from sklearn.kernel_ridge import KernelRidge
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV

# define pipeline
pipeline = make_pipeline(KernelRidge())

# define parameter grid for hyperparameter tuning
param_grid = {'kernelridge__kernel': ['linear', 'rbf', 'poly'],
              'kernelridge__alpha': [0.1, 1, 10],
              'kernelridge__gamma': [0.1, 1, 10]}

# perform grid search
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# print best parameters and best score
print("Best parameters: ", grid_search.best_params_)
print("Best score: ", -grid_search.best_score_)

# make predictions on test set
y_pred = grid_search.predict(X_test)

# evaluate model performance using R^2 and RMSE
from sklearn.metrics import r2_score, mean_squared_error
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("R^2: ", r2)
print("RMSE: ", rmse)
```

```python
#Kernel regression even worst than linear rgeression
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso


lasso = Lasso()

param_grid = {'alpha': np.logspace(-4, 1, 6)}

# Use GridSearchCV to find the best alpha value
lasso_cv = GridSearchCV(lasso, param_grid, cv=10)
lasso_cv.fit(X_train, y_train)

# Print the best alpha value
print("Best alpha: ", lasso_cv.best_params_)

# Fit the model using the best alpha value
lasso = Lasso(alpha=lasso_cv.best_params_['alpha'])
lasso.fit(X_train, y_train)

# Predict the test set results
y_pred = lasso.predict(X_test)

# Calculate the root mean squared error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error:", rmse)

# Calculate the R-squared value
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)

# Visualize the actual and predicted values
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Yield and Energy')
plt.ylabel('Predicted Yield and Energy')
plt.title('Actual vs Predicted Values')
plt.show()
```

```python
#It is still the worst model not able to explain any actual data points through the model
#It may be due to the data points i have selected
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge


# Define the model
ridge = Ridge()

# Define the parameter grid to search
param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}

# Create the grid search object
grid_search = GridSearchCV(ridge, param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the grid search to the training data
grid_search.fit(X_train, y_train)

# Print the best parameters and score
print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", -grid_search.best_score_)

# Use the best parameters to fit the model to the training data
ridge_best = Ridge(alpha=grid_search.best_params_['alpha'])
ridge_best.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = ridge_best.predict(X_test)

# Calculate the RMSE and R^2 score
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

# Print the RMSE and R^2 score
print("RMSE: ", rmse)
print("R^2 Score: ", r2)
```

```python
# Visualize the actual and predicted values
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Yield and Energy')
plt.ylabel('Predicted Yield and Energy')
plt.title('Actual vs Predicted Values')
plt.show()
```

```python
#It is still the worst model not able to explain many actual data points through the model
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

# Create Random Forest model
rf_model = RandomForestRegressor(random_state=42)

# Train the model
rf_model.fit(X_train, y_train)

# Predict on test set
y_pred = rf_model.predict(X_test)

# Evaluate the model using RMSE and R^2
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)
print('RMSE:', rmse)
print('R^2:', r2)
```

```python
# Visualize the actual and predicted values
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Yield and Energy')
plt.ylabel('Predicted Yield and Energy')
plt.title('Actual vs Predicted Values')
plt.show()
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

# creating the decision tree regressor model
tree_model = DecisionTreeRegressor(random_state=42)

# train the model
tree_model.fit(X_train, y_train)

# make predictions on the test data
y_pred = tree_model.predict(X_test)

# evaluate the model using r2 score and mean squared error
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print("R2 score: ", r2)
print("MSE: ", mse)
```

```python
# Visualize the actual and predicted values
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Yield and Energy')
plt.ylabel('Predicted Yield and Energy')
plt.title('Actual vs Predicted Values')
plt.show()
```

```python
#this one improved a bit further with even better R2 value
```

# K-NN

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score

k_range = range(1, 31)
k_scores = []

for k in k_range:
    knn = KNeighborsRegressor(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='r2')
    k_scores.append(scores.mean())

# find the best value of k
best_k = k_range[np.argmax(k_scores)]

# predict on test set using the best k
knn = KNeighborsRegressor(n_neighbors=best_k)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# calculate R2 score
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)

print("Best value of k: ", best_k)
print("R2 score on test set: ", r2)
```

```python
# Visualize the actual and predicted values
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Yield and Energy')
plt.ylabel('Predicted Yield and Energy')
plt.title('Actual vs Predicted Values')
plt.show()
```

```python
#K-NN is the best models among all the models i have used
```
